```python
# Import SQLAlchemy `automap` and other dependencies
from sqlalchemy import create_engine, inspect
```

```python
# Create the connection engine
engine = create_engine("sqlite:///bellybutton.sqlite")
conn = engine.connect()
```

```python
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
from sqlalchemy import Column, Integer, String, Float

# class Otu(Base):
#     __tablename__ = "otu"
#     otu_id = Column(Integer, primary_key=True)
#     otu_label = Column(String(1000))
#     s940 = Column(Integer)
    
    

class Metadata(Base):
    __tablename__ = "samples_metadata"

    sample = Column(Integer, primary_key=True)
    EVENT = Column(String)
    ETHNICITY = Column(String)
    GENDER = Column(String)
    AGE = Column(Integer)
    WFREQ = Column(Float)
    BBTYPE = Column(String)
    LOCATION = Column(String)
    COUNTRY012 = Column(String)
    ZIP012 = Column(String)
    COUNTRY1319 = Column(String)
    ZIP1319 = Column(String)
    DOG = Column(String)
    CAT = Column(String)
    IMPSURFACE013 = Column(Integer)
    NPP013 = Column(Float)
    MMAXTEMP013 = Column(Float)
    PFC013 = Column(Float)
    IMPSURFACE1319 = Column(Integer)
    NPP1319 = Column(Float)
    MMAXTEMP1319 = Column(Float)
    PFC1319 = Column(Float)
```

```python
Base.metadata.create_all(conn)
```

```python
df.to_sql(Metadata.__tablename__, conn, index=False, if_exists="replace")
```
